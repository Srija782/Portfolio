```python
# Finding Correlations
#scatter plot with budget vs gross revenue

plt.scatter(x=df['budget'],y=df['gross'])
plt.title("Budget Vs Gross")
plt.xlabel("Budget")
plt.ylabel("Gross")
plt.show()
```

```python
df.corr(method= "pearson") #correlation is for only numerical vars, methods = pearson,kendall and spearman
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
df_numerised = df
for col in df_numerised.columns:
    if(df_numerised[col].dtype=='object'):
        some_map = {label:idx for idx,label in enumerate(np.unique(df_numerised[col]))} #create a map w.r.t the category column
        df_numerised[col] = df_numerised[col].map(some_map )
```

```python
df_numerised.sort_values(by='gross',inplace=False,ascending=False)
```

```python
sns.heatmap(df_numerised.corr(),annot=True)
```

```python
df_numerised.corr()
```

```python
# Unstacking the correlation matrix

corr_matrix = df_numerised.corr()
corr_pairs = corr_matrix.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr=sorted_pairs[sorted_pairs>0.5]
high_corr
```
